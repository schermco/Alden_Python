```python
fipsi_data = pd.read_csv("C:\\Users\\gdsch\\Documents\\Fipsi_temperature_log.csv",sep = ',')
fipsi_data.head(1000)
```

```python
from pandas import DataFrame

df = DataFrame(fipsi_data,columns=['Date_time', 'Temperature'])

df.loc[df.Temperature > 80, 'Alert'] = 'Alert! Fipsi is too hot!'  

df.loc[df.Temperature < 70, 'Alert'] = 'Alert! Fipsi is too cold!'  

df.loc[(df.Temperature <= 80) & (df.Temperature >= 70), 'Alert'] = 'Fipsi is comfortable'

print(df)
```
